```
def func_investidor(spark, targetDirectory):
  
  jdbcURL = sc._jvm.com.databricks.training.ClassroomDB.jdbcURL(spark._jsparkSession)
  dbProps = dict(sc._jvm.java.util.HashMap(sc._jvm.com.databricks.training.ClassroomDB.dbProps()))
  
  
  # Conecte-se ao Banco de Dados por meio de JDBC e leia a tabela de investidores em um DataFrame.
  dfInvestidor = (spark.read
                       .jdbc(url         = jdbcURL,           # JDBC URL
                             table       = "investors",       # nome da tabela
                             properties  = dbProps))          # propriedades da conexão 
  
  
  # Usando a função `.show()` para exibir os primeiros 10 registros do DataFrame dos investidores.
  dfInvestidor.show(10, truncate = False)

  
  # Grave o DataFrame como uma tabela Delta para targetDirectory.
  # Certifique-se de deixar o Spark sobrescrever o local de destino, se necessário.
  (dfInvestidor.write
               .mode("overwrite")
               .format("delta")
               .save(targetDirectory))
  
  
  return dfInvestidor
```

```
dfFinal = func_investidor(spark, targetDirectory)
```
